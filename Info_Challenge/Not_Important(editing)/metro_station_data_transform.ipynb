```python
#make it a single dict
station_type_dict = {}
for station in station_types:
    for name, types in station.items():
        station_type_dict[name] = types

station_types_list = []

for index, row in df.iterrows():
    station_name = row['NAME']
    station_types_list.append(station_type_dict.get(station_name, []))  

df['STATION_TYPE'] = station_types_list

print(df)
```

```python
#add NUMBER_OF_LINES
def count_lines(line_str):
    return len(line_str.split(','))

df['NUMBER_OF_LINES'] = df['LINE'].apply(count_lines)

print(df)
```
